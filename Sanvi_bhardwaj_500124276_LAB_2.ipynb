```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('diabetes.csv.zip')
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna()
```

```python
numerical_columns = data.select_dtypes(include=['float']).columns
```

```python
numerical_columns
```

```python
for column in numerical_columns:
  plt.figure(figsize=(8, 6))
  sns.histplot(data[column], kde=True) # Change 'true' to 'True'
  plt.title(f"histogram of {column}")
  plt.xlabel(column)
  plt.ylabel("frequency")
  plt.show()
```

```python
print(numerical_columns)
```

```python
if len(numerical_columns) >= 2:
  plt.figure(figsize=(6, 4))
  sns.scatterplot(data = data, x=numerical_columns[3], y=numerical_columns[2] )
  plt.title(f"scatter plot : {numerical_columns[3]} vs { numerical_columns[2]}")
  plt.xlabel(numerical_columns[3]) # Corrected from plt.xlable to plt.xlabel
  plt.ylabel(numerical_columns[2]) # Corrected from plt.xlable to plt.ylabel
  plt.show()
```

```python
weight = data['weight']
cholesterol = data['chol']

plt.figure(figsize=(8, 6))
sns.scatterplot(x=weight, y=cholesterol, color='skyblue')
plt.title('graph plotted between weight and Cholesterol', fontsize=14)
plt.xlabel('weight', fontsize=12)
plt.ylabel('Cholesterol', fontsize=12)
plt.grid(True, alpha=0.5)

plt.legend()
plt.tight_layout()
plt.show()
```
